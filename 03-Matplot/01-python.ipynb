```python
# Create a line plot

x = [1,2,3,4,5]
y = [2,4,6,78,0]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Single Line Plot')
plt.show()
```

```python
# Customized Line Plot
    # linestyle: -, ., --, -.
    # marker: o, *

x = [1,2,3,4,5]
y = [2,4,6,78,0]

plt.plot(x,y, color='green', linestyle='--', marker='o', linewidth=0.5, markersize=7)
plt.grid(True)
```
